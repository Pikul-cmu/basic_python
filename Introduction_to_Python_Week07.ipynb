<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2"><code style="color:inherit">while</code> Statement</h1>

The `while` loop in Python is a fundamental control flow statement.  It allows you to repeatedly execute a block of code as long as a specified condition remains `True`. Unlike a `for` loop, which iterates over a sequence (like a list or string), a `while` loop continues as long as a boolean expression evaluates to `True`.




<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Basic <code style="color:inherit">while</code> Loop Syntax</h1>



```python
# Basic syntax
while condition:
    # Code block to be executed repeatedly
    # as long as 'condition' is True
    # ...
    statement1
    statement2
# Code to be executed AFTER the loop finishes
statement3
```

- **condition**: A boolean expression (something that evaluates to True or False). This could be a variable, a comparison (e.g., x < 10), a function call that returns a boolean, or a combination of these using logical operators (and, or, not).

- **: (colon)**: Required at the end of the while statement.

- **Indentation**: The code block inside the loop must be indented (typically with 4 spaces). This indentation tells Python which statements belong to the loop. This is critical in Python.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Simple <code style="color:inherit">while</code> Loop Examples</h1>

````python
# Example 1: Calculate the number from 1 t0 100; 1+2+3+ . . . + 100
sum = 0
num = 1
while num < 101:
    sum += num
    num += 1  # IMPORTANT: Increment the number!
print(f'''Loop finished!
Sum = {sum}''')

````

The above program works equivalently to:

````python
sum = 0
for num in range(1,101):
  sum+= num
print(f'''Loop finished!
Sum = {sum}''')
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2"><code style="color:inherit">break</code> and <code style="color:inherit">continue</code> Statements </h1>

- **`break`**: Immediately exits the entire `while` loop, regardless of whether the condition is still `True`. Control jumps to the statement immediately after the loop.
- **`continue`**: Skips the rest of the current iteration of the loop and goes back to the beginning of the loop, re-evaluating the condition.


````python
# Example of 'break':  Stop when a specific number is found
numbers = input(f"insert sequence number:").split(",")
numbers = [int(x) for x in numbers]
index = 0
while index < len(numbers):
    if numbers[index] == 8:
        print("Found 8 at index", index)
        print()
        break  # Exit the loop immediately
    index += 1
print("----- next example ------- ")

# Example of 'continue':  Print only odd numbers
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:  # If the number is even
        continue       # Skip the rest of this iteration
    print(count)       # This line will only execute for odd numbers
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">The <code style="color:inherit">else</code> Clause with <code style="color:inherit">while</code> Loops </h1>

Python has a unique feature: a `while` loop can have an optional `else` clause.  The `else` block executes only if the loop completes normally (i.e., the condition becomes `False`).  If the loop is exited prematurely with a `break` statement, the `else` block is skipped.


````python
# Example: Search for a number in a list, using 'else'
numbers = input(f"insert sequence number: ").split(",")
numbers = [int(x) for x in numbers]

target = int(input(f"insert the number you want to check "))
index = 0

while index < len(numbers):
    if numbers[index] == target:
        print(f"Found {target} at index {index}")
        break  # Exit the loop, so the 'else' block won't run
    index += 1
else:
    # This block runs only if the loop completes without a 'break'
    print(f"{target} not found in the list")

````

The `else` clause is particularly useful in search scenarios.  It lets you cleanly handle the case where the item you're searching for isn't found, without needing extra flags or variables.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Nested <code style="color:inherit">while </code>Loops</h1>

You can have `while` loops inside other `while` loops.  This is called nesting.  Be careful with indentation!

````python
# Example: Print a multiplication table (up to 12x12)
row = 1
while row <= 12:
    col = 1
    print(f"Multiplication table of {row}:",end = "")
    while col <= 12:
        product = row * col
        print(f"{product:5}", end="")  # Print with formatting
        col += 1
    print()  # Newline after each row
    row += 1
````

In nested loops, a `break` or `continue` statement only affects the innermost loop it's contained within.
