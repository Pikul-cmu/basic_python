String Data Type and Operators</h1>

As mentioned, strings in Python start and end with a **single quotes** (`'`) or **double quotes** (`"`). ***Strings always start and end with the same style of quote. You can not mix the type of quotes.***

A string can be made up of letters, numbers, and special characters.

```python
print('What is Python?')
print("Python 2 vs. Python 3")
```

To display single quotes **`'`** in a string, you can use double quotes as the outer quotes:

```python
print("It's time to learn Python")
```

To display double quotes **`"`** in a string, you can use single quotes as the outer quotes:

```python
print('I said "Hi" to you')
```

### Escape sequences

If you use the ***same type of quote within your string, the Syntax Error will occur due to the confusion between the type of quote.***

At this moment, the qoute inside the string become ***an illegal character***.

To solve problem, we need to <em style="color:blue">escape</em> the quote by inserting the `\` (backslash) which is an <em style="color:blue">escape character</em> in front of the illegal character.

```python
print('It 's time to learn Python')
```

```python
print('It\'s time to learn Python')
```

```python
print("I said \"Hi\" to you")
```

`\'` and `\"` are examples of what we call an escape sequence.

An <em style="color:blue">escape sequence</em> begins with an escape character (`\`), which indicates that the character(s) that follow should be treated in a special way.

Escape sequences can be used to display the following characters:
  - `\\` &nbsp; backslash (`\`)  
  - `\'` &nbsp; single quote (`'`)  
  - `\"` &nbsp; double quote (`"`)  
  - `\t` &nbsp; tab
  - `\n` &nbsp; return or newline

We use escape sequences in strings - usually with `print()` function.

### Syntax error on Multiline string

If you create a string using single or double quotes, the whole string must fit onto a single line.

Here's what happens when you try to stretch a string accross multiple line:

```python
print('hello
world')
```

To span multiple lines, we need to put three single quotes or three double quotes around the string instead of one of each. **Triple quotes are useful for creating strings containing paragraphs or multiple lines.**

```python
print('''hello
world''')
```

A nice benefit of using triple-quoted strings is that you can embed single and double quotes inside it without escaping them:

```python
print('''This string has single ' and " quotes inside of it''')
```

> Note:

Numeric Data Types and Operators</h1>

### Two of Python's numeric types:

- **Integer**  
    - e.g., 3, 4, 894, 0, -3, -18  
- **Floating point number** (an approximation to a real number)
    - e.g., 5.6, 7.342, 53452.0, 0.0, -89.34, -9.5


### Arithmetic operators
            
| Operator | Operation         | Expression | English Description | Result |
|----------|-------------------|------------|---------------------|--------|
|  +       | addition          | 11 + 56    | 11 plus 56          | 67     |
|  -       | subtraction       | 23 - 52    | 23 minus 5          | -29    |
|  *       | multiplication    | 4 * 5      | 4 multiplied by 5   | 20     |
|  **      | exponentiation    | 2 ** 5     | 2 to the power of 5 |	32     |
|  /       | division	       | 9 / 2      | 9 divided by 2      |	4.5    |
|  //      | integer division  | 9 // 2     | 9 divided by 2      |	4      |
|  %       | modulo (remainder)| 9 % 2      | 9 mod 2             |	1      |


### Arithmetic operator precedence
            
- When multiple operators are combined in a single expression, the operations are evaluated in order of precedence.
- All operators except the `**` operator **are evaluated from left to right** and are listed in the following table from lowest to highest precedence.
- The parentheses can be used to group code to achieve the precedence you want.

| Operator     | Description |
|--------------|-------------|
| ( )          | Parentheses|
| **           | Exponentiation|
| *, /, //, %  | Multiplication, division, integer division, modulo|
| +, -         | Addition, subtraction|


### Notes on numeric and string data types

- Numbers are **not placed in quotes**      
- A string can contain any character including **numbers**, but it is **placed in quotes**

```python
# 20 is int and ' students are signed up for Python class' is str
print(20 + ' students are signed up for Python class')  # Error due to different data types with additive operator.
```

- This code causes an error because Python only allows combining *like types*:

    - **number** + **number**
