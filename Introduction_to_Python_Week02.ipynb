Variables</h1>

- When coding, we often want Python to <em style="color:blue">remember</em> some value for later use.
- We need to pick a name for that value to be remembered and we can use that name to retrieve the value later.
- We use the term <em style="color:blue">variable</em> to refer to the names we use to refer to this stored data. **You can think of a variable as a name, label or tag.**
- The assignment operator (`=`) simply creates an association between a name and a value.

````python
cookie = 30 * 3
print(cookie)
````

````python
juice = 60 * 5
print(juice)
````

````python
print(cookie + juice)
````

````python
greeting = 'Hello'
name = 'Joe'
print(greeting + ', ' + name)
````

### Variables can be assigned to any type of data.  

Although each value has an associated type such as an integer or string, **variable names are untyped** and can be made to refer to any type of data during execution.

```python
actor = 'Chris Hemsworth'     # A string, Chris Hemsworth, is assigned to variable "actor"
inventory = 1000              # An integer, 1000, is assigned to variable "inventory"
weight = 62.5                 # A floating point number, 62.5, is assigned to variable "weight"
```

Variable names can be bound to different values, possibly of varying types, during program execution.

```python
x = 37
x = 'Hello World'
```  

Python also allows single/multiple assignment to multiple variables at the same time.

``` python
# Assigned single value to multiple variables.
x = y = z = 28             #An integer, 28, is assigned to variable "x", "y", and "z".
print(x,y,z)
```

``` python
#Assigned multiple values to multiple variables.
x, y, z = 28, 1.22,"Hello world"            #An integer, 28, is assigned to variable "x".
                                            #A floating point, 1.22, is assigned to variable "y".
                                            #A string, Hello world, is assigned to variable "z".
print(x,y,z)
```


## Variable name

**The rules for legal Python names:**
1. Names must start with a letter (A-Za-z) or _ (underscore).
2. Names cannot start with a number.
3. Names can contain only letters, digits, and _: A-Z, a-z, 0-9, and _.
4. Names cannot contain space or special character ***(except, '_' ,underscores)***.
5. Names must not be a Python keyword.


Types and Objects</h1>

In assignment, we associate a value with a variable. What exactly is that value?

Every piece of data stored in a program is an object. Each object has an identity, a value, and a type. For example, when you write:

```python
x = 37
```
- An integer object is created with the value of `37`.
- The identity of an object is the actual memory address of the object.
- The variable `x` contains the memory address of the object to which it refers.
- The `type` determines both what an object can be and what can be done to it.   
    - For example, in type `int`, the values are `..., -3, -2, -1, 0, 1, 2, 3, ...` and we have seen that these operators can be applied to those values: `+, -, *, /, //, %, **`.

Formatting Strings using the placeholders</h1>

- When creating strings to present as messages to the user of a program, you will want to format them in such a way that they are clear and easy to understand.
- While it is possible to format strings using concatenation (`+`) or text repeatation (`*`), this can quickly become tedious.
- Fortunately, Python provides a way to set up the format of string called as **" the Placeholder "**
- Placeholder is a special symbol or **construct** within a string **that marks a position where a value from the output will be inserted later**. Think of it like a blank space in a form that you fill in with specific information.

Python offers several ways to use placeholders for string formatting. Let's explore the most common methods:

## `%`-formatting (Old-style)

This is the oldest method, similar to how string formatting is done in C. You use special characters (placeholders) following with a string formatting operator (`%`) to indicate where values should be inserted.

**Placeholders:**
 - `%s`: String
 - `%d`: Integer
 - `%f`: Floating-point number (can control up to 6 digit of the floating point)

To use a placeholder, we put the **Placeholders** inside a fixed string
to hold places for variables.

After the string, insert the string formatting operator (`%`) (not a comma) following with a variable or a list of variables.

**Note that**: The number of placeholders must match with the number of variables.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#B24C00">
Example</h1>

````python
name = "Alice"
print("My name is %s." % (name))  # Output: My name is Alice.
````

````python
name = "Alice"
age = 30
print("My name is %s and I am %d years old." % (name, age))  # Output: My name is Alice and I am 30 years old.
````

````python
x = 16; y = 26.42; z = 7.569091
print("x = %d, y = %.1f, z = %.2f" %(x,y,z))
````

````python
# What happens if the number of variables does not match the number of placeholders?
x = 16; y = 26.42; z = 7.569091
print("x = %d, y = %.1f, z = %.2f" %(x,y))
````

 ## `f-strings` format (Modern-style)

 f-strings offer the most concise and readable way to format strings. You prefix the string with `f` and embed expressions directly inside curly braces `{}`.

 **Placeholders**: Expressions within curly braces `{}`.

 <h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#B24C00">
Example</h1>

````python
name = 'john'
print(f'My name is {name}.') # Output: My name is John.
````

Python will look in the placeholder and evaluate the code there. Note that the placeholder can contain function calls, method calls, or any other arbitrary code:

````python
age = int(input('How old are you? '))
print(f'Next ten year, you will be {age + 10} year old')
````

````python
number = 25
print(f'Square root of {number} is {number**.5}')
````

Input-Process-Output Pattern</h1>

Input-Process-Output (IPO) pattern is one of basic program structures. It
starts from accepting a number of inputs from a user. Then, it performs a
number of operations to generate an output. Finally, the output is displayed to
the user.

Input-Process-Output Pattern.JPG

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#B24C00">
Example</h1>

1) Write a program that accepts the width (*`w`* ) and the height (*`h`*) of a rectangle. Then, calculate and displays the area (*`a`*) of the rectangle from the formula below.

$$a = w \times h$$

````python
# INPUT
width = input("Enter the width: ")
height = input("Enter the height: ")
# PROCESS
width = float(width)
height = float(height)
area = width*height
# OUTPUT
print(f"The area of the rectangle = {area}")
````

2) Write a program that accepts a temperature in kelvin (*`k`*) from a user. The program then converts the temperature to degree celsius and displays the result with 2 digits after the decimal point.

$$c = k - 273.15$$

````python
# INPUT
kelvin = input("Input the temperature in kelvin: ")
# PROCESS
kelvin = float(kelvin)
celsius = kelvin - 273.15
# OUTPUT
print(f"Temperature in degree celsius: {celsius}")
````

3) Write a program that accepts a temperature in degree celsius (*`c`*). It then converts the temperature to degree fahrenheit (*`f`*) and displays the result with 2 digits after the decimal point.

$$f = \frac{9c}{5} + 32$$

````python
# INPUT
celsius = input("Input the temperature in degree celsius: ")
# PROCESS
celsius = float(celsius)
fahrenhit = (9*celsius/5) + 32
# OUTPUT
print(f"Temperature in degree fahrenhit: {fahrenhit:.5f}")
````

4) The **stone, pound,** and **ounce** are English units of mass, where
- 1 stone is equal to 14 pounds
- 1 pound is equal to 16 ounces.
- 1 kilogram is equal to 2.2 pounds.

Write a program that accepts a mass in **stones, pounds, and ounces** from a user. It then calculates and displays the equivalent mass in **kilograms**.

````python
# INPUT
stone,pound,ounce = input("Input a mass in stones, pounds, ounces:").split()
# PROCESS
stone,pound,ounce = float(stone), float(pound), float(ounce)
kilograms = ((stone*14) + pound + (ounce/16))/2.2
#OUTPUT
print(f"The equivalent mass in kg = {kilograms:.2f}")
````
