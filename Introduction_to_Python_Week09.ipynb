<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1. Functions </h1>

Functions are essential building blocks in Python programming. They allow you to group a set of statements together to perform a specific task. But why go to the trouble of creating them? Let's explore the importance of functions.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.1 Why Use Functions? The calculating areas</h1>
Imagine you need to write a program to calculate the areas of different shapes, say rectangles and circles. You might need to do this calculation multiple times for different dimensions.

**Scenario 1: No Functions (Repeating Yourself)**

Without functions, every time you need to calculate an area, you write the formula directly in your main code:

1.  Set rectangle 1 width = 5, height = 10.
2.  Calculate area1 = 5 * 10.
3.  Print "The area of the rectangle is [area1]".
4.  Set circle 1 radius = 7.
5.  Calculate area2 = 3.14159 * 7 * 7.
6.  Print "The area of the circle is [area2]".
7.  Set rectangle 2 width = 3, height = 8.
8.  Calculate area3 = 3 * 8.
9.  Print "The area of the rectangle is [area3]".
10. Set circle 2 radius = 4.
11. Calculate area4 = 3.14159 * 4 * 4.
12. Print "The area of the circle is [area4]".

Notice the repetition? The logic for calculating the rectangle area (width * height) is written out twice. The logic for the circle area (pi * radius^2) is also written out twice. If you made a mistake in the formula (e.g., forgot pi) or wanted to use a more precise value for pi, you'd have to find and fix it in multiple places.

**Scenario 2: Using Functions (Reusable Logic with Arguments)**

Now, let's create functions for these calculations. Crucially, these functions will accept *arguments* (like width, height, or radius) so they can work for different shapes.

1.  Define `calculate_rectangle_area(width, height)`: Takes width and height, returns area.
2.  Define `calculate_circle_area(radius)`: Takes radius, returns area.
3.  Define `display_area(shape_name, area)`: Takes the shape's name and its calculated area, and prints it nicely.

Your process now looks like this:

1.  `rect_area1 = calculate_rectangle_area(5, 10)`
2.  `display_area("Rectangle 1", rect_area1)`
3.  `circ_area1 = calculate_circle_area(7)`
4.  `display_area("Circle 1", circ_area1)`
5.  `rect_area2 = calculate_rectangle_area(3, 8)`
6.  `display_area("Rectangle 2", rect_area2)`
7.  `circ_area2 = calculate_circle_area(4)`
8.  `display_area("Circle 2", circ_area2)`

**This approach highlights the advantages**:

1.  **Reusability (DRY):** The *logic* for calculating rectangle area is written once inside `calculate_rectangle_area`. We reuse it by calling the function with different *arguments* (5, 10) and (3, 8). The same applies to `calculate_circle_area`.
2.  **Organization & Modularity:** The code is neatly organized into logical units. One function handles rectangle areas, another handles circle areas, and a third handles displaying the results.
3.  **Readability:** The main part of the code clearly shows *what* is being done (calculating and displaying areas) without getting bogged down in the *how* (the specific formulas, which are inside the functions). The function names `calculate_rectangle_area` and `calculate_circle_area` are self-explanatory.
4.  **Maintainability:** If you need to change how circle area is calculated (e.g., use `math.pi` for more precision), you only need to modify the `calculate_circle_area` function. All the places that call this function will benefit from the change without needing individual edits.


This area calculation example clearly shows how functions, especially when designed to accept arguments, allow you to write flexible, reusable, and much more maintainable code. **You define the calculation logic once and then apply it wherever needed just by calling the function with the appropriate inputs.**

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
import math # Import the math module to use math.pi

# ===== Scenario 1: Without Functions (Repetitive) =====

print("--- Calculating Areas (No Functions) ---")
# Rectangle 1
rect1_width = 5
rect1_height = 10
rect1_area = rect1_width * rect1_height
print(f"The area of Rectangle 1 ({rect1_width}x{rect1_height}) is {rect1_area}")

# Circle 1
circ1_radius = 7
# Using math.pi for consistency, but imagine typing 3.14159... manually
circ1_area = math.pi * circ1_radius * circ1_radius
print(f"The area of Circle 1 (radius {circ1_radius}) is {circ1_area:.2f}") # Format to 2 decimal places

# Rectangle 2 (Repetition of logic!)
rect2_width = 3
rect2_height = 8
rect2_area = rect2_width * rect2_height
print(f"The area of Rectangle 2 ({rect2_width}x{rect2_height}) is {rect2_area}")

# Circle 2 (Repetition of logic!)
circ2_radius = 4
circ2_area = math.pi * circ2_radius * circ2_radius
print(f"The area of Circle 2 (radius {circ2_radius}) is {circ2_area:.2f}")
````

````python
# ===== Scenario 2: With Functions (Reusable & Organized) =====

def calculate_rectangle_area(width, height):
  #"""Calculates the area of a rectangle."""
  area = width * height
  return area

def calculate_circle_area(radius):
  #"""Calculates the area of a circle."""
  area = math.pi * radius * radius
  return area

def display_area(shape_name, area, decimals=2):
   #"""Displays the calculated area for a shape."""
   print(f"The area of {shape_name} is {area:.{decimals}f}")



# --- Calculate areas using functions ---
print("--- Calculating Areas (With Functions) ---")

# Rectangle 1
area1 = calculate_rectangle_area(5, 10) # Pass arguments 5 and 10
display_area("Rectangle 1 (5x10)", area1, decimals=0) # Pass name and result

# Circle 1
area2 = calculate_circle_area(7) # Pass argument 7
display_area("Circle 1 (radius 7)", area2) # Default decimals=2

# Rectangle 2 (REUSING the function with DIFFERENT arguments)
area3 = calculate_rectangle_area(3, 8) # Pass arguments 3 and 8
display_area("Rectangle 2 (3x8)", area3, decimals=0)

# Circle 2 (REUSING the function with a DIFFERENT argument)
area4 = calculate_circle_area(4) # Pass argument 4
display_area("Circle 2 (radius 4)", area4)
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.2  Defining a Function</h1>

You define a function using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. Any input parameters or arguments should be placed within these parentheses. The code block within every function starts after the colon and must be indented. The `return` statement exits a function, optionally passing back a value to the caller. If no value is specified, `None` is returned.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Syntax</h1>

```python
def functionName(parameter1, ..., parameterN):
  statement1
  statement2
  statement3
  # Optional return statement
  return value
```

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.3 Passing Values to Functions (Parameters)</h1>

You pass values (arguments) to functions, which are received as parameters inside the function definition. By default, arguments are passed positionally, meaning they must match the order of parameters in the definition.

- **Required Parameters**: These must be provided in the correct positional order when calling the function, and the number of arguments must match exactly with the number of parameters in the function definition.
Example: If `Loves(A, B)` is defined, calling `Loves("Kitty")` or `Loves("Alice", "Bob", "Cathy")` will cause errors because the number of arguments doesn't match.

- **Default Parameters**: You can assign a default value to a parameter in the function definition. If an argument for that parameter is not provided during the function call, the default value is used

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#10 - Passing Values to Functions (Reguired parameters):</h1>

````python
# Define Loves function
def Loves(A, B):
  print(A, "loves", B)

# Call Loves function with exact number of argument.
Loves("Kitty", "Doraemon")

# Call Loves function with only 1 argument.
Loves("Kitty")   #caurse TypeError!

# Call Loves function with 3 arguments.
Loves("Kitty","Doraemon","Ultraman")   #caurse TypeError!
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#11 - Passing Values to Functions (Default parameters):</h1>

````python
# Define OddSum function with a default parameter
def OddSum(n=5): # n defaults to 5 if not provided
  sumval = 0
  i = 1
  while i <= n:
    sumval += 2 * i - 1
    i += 1
  print("The sum of first", n, "odd numbers is", sumval)

# Call OddSum
OddSum()     # Uses default n=5
OddSum(3)    # Uses n=3
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.4  Returning Values from a Function</h1>

The `return` statement is used to send a value back from the function to the part of the code that called it. A function can return `None` (implicitly or explicitly with `return None`) or one or more values. If a function returns a value, you usually assign it to a variable when you call the function. A function can return multiple values.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#12: Return value from a Function</h1>

````python
def function1(a, b):
  c = a + b / 2
  return c # Returns a single value

def function2(m, n):
  k = 3 * m + n * m
  return k # Returns a single value

# Call functions and store returned values
x = function1(2, 6) # x becomes 5.0
y = function2(2, 5) # y becomes 16
z = function1(x, 4) # z becomes 7.0
w = function2(y, z) # w becomes 160

print(f"x={x}, y={y}, z={z}, w={w}")
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.5  Scope of Variables</h1>

The "scope" of a variable determines where in your program you can access it. There are two basic scopes in Python: Global and Local.

- **Global Variables**: Declared outside of any function. They can be accessed from anywhere in the program, including inside functions.

- **Local Variables**: Declared inside a function (including parameters). They can only be accessed within that specific function. Variables defined inside a function do not clash with variables outside the function, even if they have the same name.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#13: Global Variables</h1>

````python
# Example of Global Variable
n = 10 # Global variable

def penny():
  # Can access global 'n'
  print("Penny is", n, "years old.") # Accesses global n

def amy():
  # Can access global 'n'
  print("Amy is", n + 3, "years old.") # Accesses global n

print("Leonard is", n, "years old.") # Accesses global n
penny()
print("Sheldon is", n+5, "years old.") # Accesses global n
amy()
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#14: Local Variables</h1>

````python
# Example of Local Variable
def penny():
  n = 10 # Local variable 'n' inside penny
  print("Penny is", n, "years old.")

# This 'n' below would be a different variable if defined globally,
# or cause an error if not defined. It cannot access penny's local 'n'.
# print("Trying to access penny's n:", n) # This would fail

penny()

# Another example with parameters as local variables
def welcome(name, place="SIIT"): # 'name' and 'place' are local to welcome
  print("Hi,", name + "!", "Welcome to", place + ".")

welcome("Sheldon")
# Trying to access 'name' or 'place' here would cause an error
#print(name) # This would fail
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.6  Global keyword</h1>

Normally, if you assign a value to a variable inside a function, Python creates a local variable. If you want to modify a *global* variable from *inside* a function, you must explicitly declare it using the `global` keyword at the beginning of the function.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example#15: Modifying Global Variable</h1>



````python
n = 15 # Global variable

def penny():
  global n # Declare intention to modify the global 'n'
  n = 10   # This now modifies the global 'n'
  print("Penny is (inside function)", n, "years old.")

def amy():
  # Accesses the current value of global 'n'
  print("Amy is", n + 3, "years old.") # Accesses global n

print("Initial n (global):", n)
amy()
print("n after amy:", n)
penny()
print("n after penny (global modified):", n)
````
