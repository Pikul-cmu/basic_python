<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2"><code style="color:inherit">for</code> Statement</h1>

Computers are often used to **automate repetitive tasks**. Repeating identical or
similar tasks without making errors is something that computers do well and people
do poorly. Because iteration is so common, Python provides several language
features to make it easier. One form of iteration in Python is the `for` statement.

#How does a loop works ?


image.png

For example, you can write a simple program that uses a `for` statement to display numbers 1 through 5 as follows:

```python
for num in [1, 2, 3, 4, 5]:
    print(num)
```

Syntax of a `for` statement:

```python
for element in collection:
    statement(s)
```

- Recall that a `collection` or `container` is a single objet that contains multiple elements that are associated with the collection object.
- Examples of collections are strings, list, tuples, etc.
- The variable name used for the `element` is **completely up to the you**.
- The variable refers to each element in the collection in turn and can be referenced inside the loop.

````python
for num in [1, 2, 3, 4, 5]:
    print(num)
````

````python
sum_list = 0

for element in [1,2,3,4,5]:
    sum_list = sum_list + element

print(sum_list)
````

Let's look at some examples that use a `for` loop to iterate over a string:

````python
# print each character in a string

for char in 'Hello World':
    print(char, end=',')
````

````python
# count the number of vowels in a string

s = 'Happy Birthday'
num_vowels = 0

for char in s:
    if char in 'aeiouAEIOU':
        num_vowels = num_vowels + 1

print(num_vowels)
````

````python
# collect the vowels in a string
vowels = ''

movie = 'Avengers: Age of Ultron'
for char in movie:
    if char in 'aeiouAEIOU':
        vowels = vowels + char

print(vowels)
````

Let's look at an example that uses a `for` loop to iterate over a list of strings:

````python
friends = ['Joseph', 'Glenn', 'Sally']

for friend in friends:
    print('Happy New Year:', friend)

print('Finished!!!')
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Built-in Class  <code style="color:inherit">range</code></h1>

Python has a built-in class called `range()` that is useful to use when you want to generate a sequence of numbers.

`range` takes up to 3 arguments:

                     `range(start,stop,step)`

- the first parameter, `start`, is a starting number of the list.
- the second parameter, `end`,  it is the end of the range, but not inclusive (up to but not including the number). Required.
- the third parameter, `step`, represent a size of increment/decrement steps. Assumed to be 1 if not provided.

**If only one argument is provided, it is assumed to be the end value!**

````python
# example 1
for i in range(5): #[0,1,2,3,4]
    print(i, end=' ')
````

````python
# example 2
for i in range(1,20): #[1,2,3,...,19]
    print(i, end=' ')
````

````python
# example 3
for i in range(1,20,1):                  #[1,2,3,...,19]
    print(i, end=' ')
````

````python
# example 4
for i in range(1,20,2):                  #[1,3,5,...,19]
    print(i, end=' ')
````

````python
# example 5
for i in range(1,20,4):                  #[1,5,9,13,17]
    print(i, end=' ')
````

````python
for i in range(20,0,-1):                 #[20,19,18,...,1]
    print(i, end= ' ')
````

````python
# Print the multiplcation table
print('muitiplcation table')
print('-' * 15)
a = int(input("insert the number you want to display multiplication:"))

# Print the numbers 1 through 10
# and their squares.
for i in range(1, 13):
    number = a*i
    print(f"{a}  x  {i} \t = {number}")
````

#Nest `for loops`

`For loops` can be nested inside ot other for loops as well. Nested loop structures can be very confusing and complex. Try to keep nested level mimimal as much as you can. If possible, avoid it.

````python
# Print the multiplcation table
print('muitiplcation table')
print('-' * 15)
a,b = input("insert the number you want to display multiplication:").split('-')
a = int(a)
b = int(b)
# Print the numbers 1 through 10
# and their squares.

for i in range(a,b+1):
    print(f"----------multiplication number: {i}------")
    for j in range(1, 13):

      number = j*i
      print(f"{i}  x  {j} \t = {number}")
````
