Input-Process-Output Pattern</h1>

Input-Process-Output (IPO) pattern is one of basic program structures. It
starts from accepting a number of inputs from a user. Then, it performs a
number of operations to generate an output. Finally, the output is displayed to
the user.

1. Input process: Getting User Input</h2>

2. Operation process</h2>

|Operator | Operation         | Associativity |
|----------|-------------------|------------|
|  (  )    | Parentheses       |              |
|  **      | Exponentiation    |  right-to-left|
|  +x,-x   | Positive, Negative  |right-to-left|
|  *, /, //, %   | Multiplication, Division, Floor division, Modulo (remainder) | left to right|

1.) Write a program that converts your weight and height from kilograms and
centimeters to pounds and feet, respectively. (1 kilogram = 2.20462262 pounds, 1
centimeter = 0.032808399 feet). The program output should look like this:

image.png

````python
#input

weight = input("Enter your weight in kg.: ")
height = input("Enter your height in cm.: ")

#process
weight = float(weight)
height = float(height)

weight_pound = weight*2.20462262
height_feet = height*0.032808399

#output
print(f"Your weight in pounds = {weight_pound:.2f}")
print(f"Your height in feet = {height_feet:.2f}")
````

### Mathematical functions

The math module provides a comprehensive collection of **mathematical functions and constants for various calculations**.  You can import the math module in a few different ways:   

-**Import the entire module**: `import math`  
-**Import the module and rename it**: `import math as m`  
-**Import specific functions and constants**: `from math import cos, pi`  
-**Import all functions and constants**: `from math import *`

The math module offers a wide range of functions, including:

- **Number-theoretic and representation functions:**

  - `math.ceil(x)`: Returns the ceiling of x, the smallest integer greater than or equal to x.
   
   *for example*,  `math.ceil(3.1) = 4`
  
  - `math.floor(x)`: Returns the floor of x, the largest integer less than or equal to x.  
   *for example*, `math.floor(3.9) = 3`

  - `math.fabs(x)`: Returns the absolute value of x.  
   *for example*,  `math.fabs(-3) = 3.0`

  - `math.factorial(x)`: Returns the factorial of x.  
   *for example*,  `math.factorial(4) = 24`  -> 1x2x3x4

- **Power and logarithmic functions:**

  - `math.exp(x)`: Returns $e^x$.
   
   *for example*,  `math.exp(10) =` $e^{10}$

  - `math.log(x[, base])`: Returns the logarithm of x to the given base (natural logarithm if the base is not specified).
   
   *for example*,  `math.log(5)` = $ln(5)$ , `math.log(5, 3)` = $log_3(5)$

  - `math.log2(x)`: Returns the base-2 logarithm of x.  
  - `math.log10(x)`: Returns the base-10 logarithm of x.  
  - `math.pow(x, y)`: Returns $x^y$. It is equivalent to `x**y`
  - `math.sqrt(x)`: Returns the square root of x, $\sqrt x$. It is equivalent to `x**0.5`

- **Trigonometric functions:**

  - `math.acos(x)`: Returns the arc cosine of x, in radians.  
  - `math.asin(x)`: Returns the arc sine of x, in radians.  
  - `math.atan(x)`: Returns the arc tangent of x, in radians.  
  - `math.cos(x)`: Returns the cosine of x radians.  
  - `math.sin(x)`: Returns the sine of x radians.  
  - `math.tan(x)`: Returns the tangent of x radians.  

- **Angular conversion:**

  - `math.degrees(x)`: Converts angle x from radians to degrees.   
  - `math.radians(x)`: Converts angle x from degrees to radians.   

- **Constants:**

  - `math.pi`: Returns the mathematical constant `π = 3.141592...`
  - `math.e`: Returns the mathematical constant `e = 2.718281...`

````python
# import math module and rename it as m
import math as m

# input
size_a, size_b = input("Enter lengths of two sides: ").split()
angle = input("Enter an angle between them: ")

#process
size_a = float(size_a)
size_b = float(size_b)
angle = m.radians(float(angle))
area = ((size_a*size_b)*m.sin(angle))/2

#output
print(f"The area of the triangle is {area:.2f}")
````

###  Random module

The `random` module is used to generate random numbers, which can be useful for various applications like games, simulations, and cryptography. You can import the random module in a few different ways:   

-**Import the entire module**: `import random` `Inline code`  
-**Import the module and rename it**: `import random as r`  
-**Import specific functions and constants**: `from random import randint`  
-**Import all functions and constants**: `from random import *`

The random module offers a wide range of functions, including:

- `random.randint(a, b)`: Returns a random integer between a and b (inclusive).

  *for example*, `random.randint(1, 10)` returns an integer between 1 and 10 (inclusive).
- `random.random()`: Returns a random floating-point number between 0.0 (inclusive) and 1.0 (exclusive).
- `random.choice(seq)`: Returns a random element from the sequence seq.
  
  *for example*, random.choice(['apple', 'banana', 'cherry']) returns a random fruit from the list.

- `random.shuffle(x)`: Shuffles the elements of the sequence x in place.
  
  *for example*, if x =, random.shuffle(x) shuffles the order of elements in the list.

###  String methods

Python provides various string methods to manipulate and work with strings effectively. Unlike `math` and `random` modules, the `str` module is directly accessible, and **you can use its methods without explicitly importing it**.

Some useful string methods include:

- `str.capitalize()`: Capitalizes the first character of the string.

  *for example*, `"hello".capitalize()` returns `"Hello"`.

- `str.lower()`: Converts the string to lowercase.

  *for example*, `"HeLlO".lower()` returns `"hello"`.

- `str.upper()`: Converts the string to uppercase.

  *for example*, `"hello".upper()` returns `"HELLO"`.

- `str.isalpha()`: Checks if the string contains only alphabetic characters.

 *for example*, `"hello".isalpha()` returns True, while `"hello123".isalpha()` and `"hello".isalpha()` return `False`.

- `str.isnumeric()`: Checks if the string contains only numeric characters.

  *for example*, `"123".isnumeric()` returns True, while `"123abc".isnumeric()` and `"123 ".isnumeric()` return `False`.

- `str.isspace()`: Checks if the string contains only whitespace characters.

   *for example*, `" ".isspace()` returns `True`, while `" a ".isspace()` returns `False`.

- `str.isupper()`: Checks if the string is in uppercase.

  *for example*, `"HELLO".isupper()` returns `True`, while `"Hello".isupper()` returns `False`.

- `str.split(sep=None)`: Splits the string into a list of substrings using the delimiter `sep`.

  *for example*, `"hello world".split()` returns `['hello', 'world']`.

- `str.strip([chars])`: Removes leading and trailing characters from the string.

  *for example*, `" hello ".strip()` returns `"hello"`.


- `str.zfill(width)`: Pad a string with zeros on the left, making it a specified width. If the string starts with a plus or minus sign, the zeros are padded after the sign.

  *for example*, `"42".zfill(5)` returns `"00042"`.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#B24C00"> Example</h1>

````python
text = "Hello, World!"

print(f"Capitalized: {text.capitalize()}")  # Output: "Hello, world!"
print(f"Lowercase: {text.lower()}")  # Output: "hello, world!"
print(f"Is alphabetic? {text.isalpha()}")  # Output: False
print(f"Split into words: {text.split()}")  # Output: ['Hello,', 'World!']
````

### Installing new libraries

While `math`, `random`, and `str` are built-in modules, many other useful libraries are not included in the standard Python library. To use these external libraries, you need to install them using a tool called `pip`.

- **Windows**: Open Command Prompt and use the command:
  
  `py -3 -m pip install [library name]`

- **macOS and Linux**: Open Terminal and use the command:
  
  `pip3 install --user [library name]`

Replace `[library name]` with the actual name of the library you want to install. For example, to install the NumPy library, you would use `pip install numpy`. Once installed, you can `import` the library into your Python code using the import statement.

### NumPy

NumPy is a powerful library for numerical computing in Python. It provides tools for working with arrays, matrices, and mathematical functions. To use NumPy, You can import the Numpy library in a few different ways:   

-**Import the entire library**: `import numpy`  
-**Import the library and rename it**: `import numpy as np`  
-**Import specific functions and constants**: `from numpy import arange`  
-**Import all functions and constants**: `from numpy import *`

Some common functions in Numpy library include:

- `np.arange(start, stop, step)`: Creates an array with evenly spaced values.

  *for example*, `np.arange(0, 1, 0.1)` creates an array `[0, 0.1, 0.2,... 0.9]`.

- `np.zeros(shape)`: Creates an array filled with zeros.

  *for example*, `np.zeros(5)` creates an array of 5 zeros `[0,0,0,0,0]`.

- `np.ones(shape)`: Creates an array filled with ones.

  *for example*, `np.ones((2, 3))` creates a 2x3 array filled with ones.

- `np.identity(n)`: Creates an identity matrix.

  *for example*, `np.identity(3)` creates a 3x3 identity matrix.

NumPy also provides element-wise operations, broadcasting, reshaping, matrix multiplication, and other linear algebra functions.




````python
# import numpy library as renamed it as "np"
import numpy as np
np.set_printoptions(precision=2, floatmode="fixed")  # sets the number of digits after the decimal point to be fixed to 2.

# input
v = np.arange(0, 1, 0.2)

# output
print(v)
print(v+2)
print(np.power(v, 2))
print(np.exp(v))
````

### Matplotlib

Matplotlib is a popular library for creating visualizations in Python. It provides functions for plotting various types of graphs, charts, and figures. To use Matplotlib, You can import the Matplotlib library in a few different ways:   

-**Import the entire library**: `import matplotlib`  
-**Import the library and rename it**: `import matplotlib as plt`  
-**Import specific functions and constants**: `from matplotlib import plot`  
-**Import all functions and constants**: `from matplotlib import *`

Some common Matplotlib functions include:

- `plt.plot(x, y, **kwargs)`: Creates a line plot. The parameter `**kwargs` represents formatting parameters such as `color`, `linewidth`, `linestyle`, etc.
- `plt.xlim(xmin, xmax)`: Sets the x-axis limits.
  *for example*, `plt.xlim(0, 10)` sets the x-axis to display values between 0 and 10.
- `plt.ylim(ymin, ymax)`: Sets the y-axis limits.
- `plt.show()`: Displays the plot.
- `plt.savefig(filename)`: Saves the plot to a file.


### Function **plot()** in matplotlib.


The syntax of `plt.plot()` is as follows:



```
plt.plot(x, y, **kwargs)
```

**Parameters:**

- `x`: A list or array of x-coordinates.
- `y`: A list or array of y-coordinates.
- `**kwargs`: (Optional) Keyword arguments specifying additional plot properties, such as `linewidth`, `linestyle`, `markersize`, etc.

**Line Properties:**

- `color` or `c`: Sets the color of the line. You can use color names like 'red', 'blue', etc., or hexadecimal color codes like '#FF0000'.
- `linewidth` or `lw`: Sets the width of the line.
- `linestyle` or ls: Sets the style of the line, such as solid, dashed, dotted, etc. You can use styles like '-', '--', ':', '-.', etc.

**Marker Properties:**

- `marker`: Sets the marker style, such as circles, squares, triangles, etc. You can use marker symbols like 'o', 's', '^', etc.
- `markersize` or ms: Sets the size of the markers.
