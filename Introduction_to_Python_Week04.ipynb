# Comparison operators
  
Adding two numbers in Python results in another number; however, if two numbers were compared (i.e. to test for equality) using **comparison operators**, the result would be of `bool`.

```python
In [1]: print( 5 == 6 )

```

# Logical operators (Boolean operator)

It is also possible to combine multiple conditions using the logial operators.


*The three basic logical operators are*: **`not`, `and`, `or`**.

## **`not`**
The operator `not` is a unary operator; it *reverse the logical state* of its operand.

|`B`|`not B`|
|---|-------|
|False| True|
|True|False|

## **`and`**
The operator `and` is a binary operator; If both the operands are `True` then resulting condition becomes `True`.

|`A`|`B`|`A and B`|
|---|---|---------|
|False|False|False|
|False|True|False|
|True|False|False|
|True|True|True|

## **`or`**
The operator`or` is a binary operator; If any of two operands is `True`, then the resulting condition becomes `True`.

|`A`|`B`|`A or B`|
|---|---|---------|
|False|False|False|
|False|True|True|
|True|False|True|
|True|True|True|

>**Note that**:
>- The Logical operator `not` has the highest precedence, followed by `and`, then `or`.
>- Similar to arithmetic operators, you can control the precedence using the parentheses operator `()` to group relational expressions.


These operators are typically used to combine relational expressions rather than the values `True` and `False`. This will allow you to construct more powerful conditional statements (i.e. `if` statements).



**Example1**, if you want to test whether a variable `x` contains a number within a certain range (i.e. 10 &leq; x &leq; 20). You can perform this test, by checking whether the number is greater than or equal to 10 and whether the number is less than or equal to 20.

````python
#Example
x = 11
print( (x >= 10) and (x <= 20) )  #output: True

#Example
x = 9
print( (x >= 10) and (x <= 20) )  #output: False

````

**Example2**, you can test whether a variable `c` contains a capital letter (i.e. 'A' &leq; c &leq; 'Z').


````python
#Example

c = 'N'
print( (c >= 'A') and (c <= 'Z') )   #Output: True

#Example
c = 'n'
print( (c >= 'A') and (c <= 'Z') )   #Output: False
````

**Example4**, you can test whether a variable `x` is outside the range $[10,20]$ or not.

````python
# Testing if x is outside the range [10, 20]
x = 11
print( (x < 10) or (x > 20) )
````

**Example5**, you can test whether a variable `x` is `(a positive value) and (odd number)` or not ?

````python
# Testing if x is a positive and odd number
x = 11
print( (x > 0) and (x % 2 != 0) )
````

`if` statement in Python allow us to decide whether or not to execute a particular statement.

image.png


*Note that* the statement(s) must be **indented** at least one space to the right of the `if` keyword and **each statement must be indented using the same number of spaces**. Traditionally, we indent it **four spaces**.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2"><code style="color:inherit">if...else</code> Statement</h1>


An `if` statement takes an action if the specified condition is `True`. If the condition is `False`, nothing is done. But what if you want to take one or more alternative actions when the condition is `False`? You can use an `if-else` statement.

An `if-else` statement decides which statements to execute based on whether
the condition is `True` or `False`.

**Syntax for the if-else statement:**
image.png

````python
# comparison operator and if
number = int(input('Enter a whole number: '))

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
````

````python
# string comparison and if
person = 'John'

if person == 'John':
    print('Welcome John!')
else:
    print("Welcome, what's your name?")
````

````python
# compound conditional
x = input('Enter an odd positive integer: ')

# Create a int object that represents the user input
num = int(x)

# Test number for validity
if ((num > 0) and (num % 2 != 0)):
    print(num, "is a valid number")
else:
    print(num, "is NOT a valid number")
````

````python
#Using conditional statements with the math module
import math

x = float(input("Enter a number: "))
if x >= 0:
    print(f"The square root of {x} is {math.sqrt(x)}")
else:
    print("Cannot calculate the square root of a negative number.")
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2"><code style="color:inherit">if..elif..else</code> Statement</h1>

An `if..elif..else` statement can be used to implement multiple alternatives.

**Syntax for the multi-way `if..elif..else` statement:**

image.png

**`elif`** means **"else, if "** another condition exists do something else. You can put in **as many `elif` clauses as you want**. The test conditions are evaluated from top to bottom until one produces `True` or until there are no test conditions left to evaluate. When a test condition produces `True`, the body associated with it is executed and then the `if` statement exits. Any subsequent test conditions are ignored.

The **`else`** clause is optional and is normally used after the last test condition.

````python
# simple welcome
person = 'George'

if person == 'John':
    print('Welcome John!')
elif person =='George':
    print('Welcome George!')
else:
    print("Welcome, what's your name?")
````

````python
# secret number guess
secret_number = '2'

guess = input('Enter a guess for the secret number (1-3): ')

if guess == '1':
    print('Guess is too low')
elif guess == secret_number:
    print('Guess is right')
elif guess == '3':
    print('Guess is too high')
else:
    print(guess, 'is not a valid guess (1-3)')
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Input Validation</h1>

It is a common pracitice that a program should warn a user when the outputs from the user are invalid. In such cases, we can use conditional statements to perform the input validation.

image.png


image.png

Let's see some examples belows:

````python
#import math module and rename as m
import math as m

#input process
a, b, c = input("Please enter the values of a,b, and c:").split()

#operation process

###Perform input validation step:
if a.isnumeric() and b.isnumeric() and c.isnumeric():
  a,b,c = float(a),float(b),float(c)
  ###Start do arithmetic operations
  if a == 0:
    print("Error: This is not a quadratic equation.")
  elif (b**2 - (4*a*c)) < 0:
    print("Error: There is not a real-number solution.")
  else:
    s = m.sqrt(b**2 - (4*a*c))
    x1,x2 = (-b + s)/(2*a), (-b - s)/(2*a)
    print(f"Solution = {x1}, {x2}")

else:
  print("Error: non-numeric coefficients!!")
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Nested <code style="color:inherit">if</code> statements <h1>

A **nested `if` statements** is an *`if` statement that is placed inside another `if` statement*.  It allows you to check for multiple conditions and execute different blocks of code based on those conditions.

**Syntax for nested `if` statements:**

image.png

image.png

````python
#input process
year = int(input("What year are you?: "))

#operation process
if year == 1:
  sex = input("Male or Female?: ")
  if sex.lower() == 'male':
    room = '209'
  else:
    room = '210'
else:
  dept = input("What department are you?: ")
  if dept.lower() == "MT" or dept.lower() == "EM":
    room = '301'
  elif dept.lower() == "IT" or dept.lower() == "CPE":
    room = '302'
  else:
    room = '305'

#output process
print(f"Your room is: {room}")
````

### It can also be written in a non nested if you design your structure well.
