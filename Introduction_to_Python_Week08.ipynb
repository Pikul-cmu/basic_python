<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1. Dictionaries </h1>

Dictionaries and sets are Python’s built-in types for storing and managing collections of values. **Dictionaries are similar to lists**, *but instead of being indexed by a range of numbers, they are indexed by keys*.

Keys can be integers or strings. Think of a dictionary as a set of **key-value pairs**. This makes dictionary keys more flexible than list indices.



<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.1 Creating a Dictionary</h1>

- An empty dictionary is created using curly braces: `{}`.
- Non-empty dictionaries are created with comma-separated `key: value` pairs inside curly braces.
- Keys are separated from values by a colon (`:`).



<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Excample 1: Creating a Dictionary

empty_dict = {}
math_constants = {'pi': 3.14, 'e': 2.71}
n_students = {1: 15, 2: 19, 3: 18}

del n_students[1] #remove key
print("Empty Dictionary:", empty_dict)
print("Math Constants:", math_constants)
print("Number of Students per Class:", n_students)

````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.2 Accessing Values</h1>

Access dictionary values by using the dictionary name followed by the key enclosed in square brackets `[]`.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 2: Accessing Values
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021}

print("Course Code:", course['code'])
print("Semester:", course['semester'])
print("Academic Year:", course['acad year'])
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.3 Modifying Values</h1>

You can modify dictionaries using the assignment operator (=):



``` python
dictionary[key] = new_value
```

- if the `key` exists, its associated `value` is updated.
- if the `key` does not exist. a new key-value pair is added to the dictionary.


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 3: Modifying Values
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021}
print("Original course:", course)

# Update existing values
course['semester'] = 1 # Update semester
course['acad year'] = 2022 # Update year

# Add a new key-value pair
course['name'] = 'Intro to Computers and Programming' # Add course name

print("Modified course:", course)
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.4 Iterating through a Dictionary</h1>

You can loop through dictionaries using `for` loops in several ways:



1.   **Iterating through `keys` (default)** - Loops through the keys directly.
2.   **Iterating through Values (`.values()`)** - Loops through the values using the .values() method.
3.   **Iterating through Key-Value Pairs (`.items()`)** - Loops through both keys and values using the .items() method.
4.   **Iterating through `keys` (`.keys()`)** - Explicitly loops through keys using the .keys() method. This method returns a view object containing the keys, which can be useful with functions like `sorted()`.  

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 4: Iterating - Keys (default)
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021}
print("\nIterating through keys (default):")
for key in course:
  print(key, '->', course[key])
````

````python
#Example 5: Iterating - Values
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021} # Redefine for clarity if running cells independently
print("\nIterating through values:")
for value in course.values():
  print(value)
````

````python
#Example 6: Iterating - Items (key-value pairs)
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021} # Redefine for clarity
print("\nIterating through items (key-value pairs):")
for key, value in course.items():
  print(key, '->', value)
````

````python
#Example 7: Iterating - Keys (using .keys())
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021} # Redefine for clarity
print("\nIterating through keys (using .keys()):")
for key in course.keys():
  print(key, '->', course[key])

print("\nIterating through sorted keys:")
for key in sorted(course.keys()): # Example of using sorted()
  print(key, '->', course[key])
````

````python
# Example 8: Counting Vowels
input_string_vowels = "ITS100 Intro to Computers and Programming"
vowel_counts = {'A': 0, 'E': 0, 'I': 0, 'O': 0, 'U': 0} # Initialize counts

for char in input_string_vowels.upper(): # Convert to upper for case-insensitivity
    if char in vowel_counts:
        vowel_counts[char] += 1

print(f"Input string: {input_string_vowels}")
print("Output (Vowel Counts):")

for vowel, count in sorted(vowel_counts.items()):
    if count > 0: # Only print vowels that appeared
      print(f"{vowel} = {count}")
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.5 Checking if a Dcitionary Contains a Key</h1>

Use the `in` operator to check for the presence of a key. It returns `True` if the key exists, `False` otherwise.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 9: Checking for Keys
course = {'code': 'ITS100', 'semester': 2, 'acad year': 2021}

print("\nChecking for keys:")
print("'code' in course:", 'code' in course)
print("'semester' in course:", 'semester' in course)
print("'year' in course:", 'year' in course) # 'year' is not a key
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">1.6 Comparing Two Dictionaries</h1>

Use the equality operator (`==`) to compare dictionaries.
Two dictionaries are equal if and only if they have the exact same keys and the values associated with each corresponding key are also equal. The order of items does not matter.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 10: Comparing Dictionaries
course1 = {'code': 'ITS100', 'semester': 2, 'acad year': 2021}
course2 = {'semester': 2, 'acad year': 2021, 'code': 'ITS100'} # Same keys/values, different order
course3 = {'code': 'ITS100', 'semester': 1, 'acad year': 2021} # Different value for 'semester'
course4 = {'code': 'ITS100', 'semester': 2} # Missing 'acad year' key

print("Comparing Dictionaries:")
print("course1 == course2:", course1 == course2) # True, order doesn't matter
print("course1 == course3:", course1 == course3) # False, different semester value
print("course1 == course4:", course1 == course4) # False, course4 missing a key
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2. Sets</h1>

A set is an unordered collection of unique values.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2.1 Creating a Set</h1>

- Define a set with comma-separated values inside curly braces `{}`.
- Create a set from a list using the `set()` built-in function. Duplicates in the list are removed, and order is not preserved.
- Create an empty set using `set()` (using `{}` creates an empty dictionary).

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 11: Creating a Set
fruits_literal = {'apple', 'orange', 'grape'} # Using literal {}
drinks_from_list = set(['water', 'milk', 'tea', 'coffee', 'tea']) # Using set() from list with duplicates
empty_set_example = set() # Creating an empty set

print("\nSet created with {}:", fruits_literal)
print("Set created from list:", drinks_from_list) # Note 'tea' appears only once, order may vary
print("Empty set:", empty_set_example)
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2.1 Adding a Value to a Set</h1>

Use the `.add()` method to add a new element.

- If the element is already present, the set remains unchanged (due to uniqueness).
- Set elements must be immutable (e.g., numbers, strings, tuples). Lists cannot be added to sets because they are mutable.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 12: Adding to a Set
fruits = set()
print("\nAdding to set:")
print("Initial set:", fruits)
fruits.add('apple')
fruits.add('watermelon')
print("After adding apple, watermelon:", fruits)
fruits.add('apple') # Adding 'apple' again has no effect
print("After adding apple again:", fruits)

# Example of trying to add a mutable type (list) - will cause an error
# try:
#     fruits.add(['lime', 'lemon'])
# except TypeError as e:
#     print(f"\nError adding list: {e}") # Lists are mutable, cannot be added

````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2.2 Removing a Value to a Set</h1>

Use the `.remove()` method to remove the element.

- If the value does not exist in the set, it raises a `KeyError`


<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 13: Removing from a Set
fruits = {'apple', 'orange', 'grape'}
print("\nRemoving from set:")
print("Initial set:", fruits)
fruits.remove('apple')
print("After removing apple:", fruits)

# Trying to remove an element that doesn't exist
try:
    fruits.remove('kiwi') # 'kiwi' is not in the set
except KeyError as e:
    print(f"Error removing kiwi: {e}") # Raises KeyError
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2.3 Checking if a Set contains a Value</h1>

Use the `in` operator, similar to lists and dictionaries.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 14: Checking Membership in a Set
fruits = {'apple', 'orange', 'grape'}
print("\nChecking membership in set:")
print("'apple' in fruits:", 'apple' in fruits)
print("'kiwi' in fruits:", 'kiwi' in fruits)
````

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">2.4 Set Operations</h1>

Sets support standard mathematical operations:
- **Union** (`|`): Returns a new set with elements from either set.
- **Intersection** (`&`): Returns a new set with elements common to both sets.
- **Difference** (`-`): Returns a new set with elements in the first set but not in the second.
- **Subset** (`<=`): Returns `True` if all elements of the first set are in the second set (including equality).
- **Proper subset** (`<`): Return `True` if all elements of the first set are in the second, but the set are not equal.

<h1 style="font-size:1.5em; font-family: verdana, Geneva, sans-serif; color:#00A0B2">Example:</h1>

````python
#Example 15: Basic Set Operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
set3 = {1, 2}

print("\nSet Operations:")
print("Set1:", set1)
print("Set2:", set2)
print("Set3:", set3)
print("Union (set1 | set2):", set1 | set2)
print("Intersection (set1 & set2):", set1 & set2)
print("Difference (set1 - set2):", set1 - set2)
print("Difference (set2 - set1):", set2 - set1)
print("Subset (set3 <= set1):", set3 <= set1)
print("Subset (set1 <= set2):", set1 <= set2)
print("Proper Subset (set3 < set1):", set3 < set1)
print("Proper Subset (set1 < set1):", set1 < set1) # A set is not a proper subset of itself
````
